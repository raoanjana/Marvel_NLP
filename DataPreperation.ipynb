```python
mc_df = df[['character', 'movie', 'line']]
```

```python
mc_df
```

```python
mc_df["character"] = mc_df["character"].apply(lambda x: x.lower())
```

```python
mc_df["movie"] = mc_df["movie"].apply(lambda x: x.lower())
```

```python
heroes =  {key.lower() : [x.lower() for x in value] for key, value in heroes.items()}
```

```python
villains =  {key.lower() : [x.lower() for x in value] for key, value in villains.items()}
```

```python
mc_df["role"] = [np.nan] * len(mc_df)
```

```python
def match_role(df, character_role_dict, role):
    characters = df["character"]
    index = 0
    for character in characters:
        if character in character_role_dict:
            df.at[index, "role"] = role
        index+=1
    return df
```

```python
mc_df_1 = match_role(mc_df, heroes, "hero")
```

```python
mc_df_2 = match_role(mc_df, villains, "villain")
```

```python
mc_df_2
```

```python
mc_df_3 = mc_df_2[mc_df_2[['role']].notna().all(axis=1)]
```

```python
def clean_lines(df):
    clean_lines = []
    for line in df["line"]:
        clean_line = clean_txt(line)
        clean_lines.append(clean_line)
    df["line"] = clean_lines
    return df
```
